```python
import os
import cv2
from ultralytics import YOLO
from tqdm import tqdm

def run_detection_inference(model, frames_dir, output_file):
    # Get list of image files sorted by frame index (assuming numeric filenames like 0.jpg, 1.jpg, ...)
    frame_files = sorted([
        f for f in os.listdir(frames_dir)
        if f.lower().endswith(('.png', '.jpg', '.jpeg'))
    ], key=lambda x: int(os.path.splitext(x)[0]))

    with open(output_file, 'w') as out:
        for frame_name in frame_files:
            frame_id = int(os.path.splitext(frame_name)[0])
            frame_path = os.path.join(frames_dir, frame_name)

            # Run detection
            results = model(frame_path)[0]

            boxes = results.boxes
            if boxes is None:
                continue

            xywh = boxes.xywh.cpu().numpy()
            confs = boxes.conf.cpu().numpy()

            for idx, (box, conf) in enumerate(zip(xywh, confs)):
                x, y, w, h = map(int, box)
                line = f"{frame_id},{idx},{x},{y},{w},{h},{conf:.4f},-1,-1,-1\n"
                out.write(line)

    print(f"Detection results saved to {output_file}")
```

```python
for data_name in os.listdir("VisDrone/VisDrone2019-MOT-test-dev/sequences"):
    run_detection_inference(model, f"VisDrone/VisDrone2019-MOT-test-dev/sequences/{data_name}", 
                                f"VisDrone/VisDrone2019-MOT-test-dev/det/{data_name}.txt")
```
